```python
git_data_path = '/home/temccrac/Programs/git_clones/ECE759_Project/cellular_clarity'
dpgp_path = '/home/temccrac/Programs/git_clones/DP_GP_cluster'
input_path = os.path.join(git_data_path, 'dpgp_input_expression.txt')
output_path = os.path.join(git_data_path, 'dpgp_output')
```

```python
# Give the parameters
a = 1 # given in authors methods as what they used for this parameter
n = 1000  # num iters (default)
os.environ["OMP_NUM_THREADS"] = "10"  # or however many threads you want
```

```python
cluster_cmd = [
    "DP_GP_cluster.py",
    "-i", input_path,
    "-o", output_path,
    "-p", "png",
    "-a", str(a),
    "-n", str(n)
]

# Initialize the system
os.chdir(dpgp_path)

# Run the program with the correct input and output data
process = subprocess.Popen(cluster_cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, bufsize=1)

# Stream the output in real-time
while True:
    output = process.stdout.readline()
    if output == '' and process.poll() is not None:
        break
    if output:
        print(output.strip())
        sys.stdout.flush()

# Capture the return code
return_code = process.poll()
if return_code != 0:
    print("Error running clustering command.")
else:
    print("Clustering completed successfully.")
```
